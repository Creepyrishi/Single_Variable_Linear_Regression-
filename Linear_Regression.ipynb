```python
import pandas as pd
```

```python
csv_data = pd.read_csv("linear_regression_dataset.csv")
```

This CSV is generated by ChatGPT with this equation Y=3X+7+noise

```python
print(csv_data)
```

```python
input_features = csv_data["X"]
output_features = csv_data["Y"]
```

```python
print(input_features)
print(output_features)
```

Now I need to add a function that will calculate error. More specifically, Squared error <br>
error = square of (predicted_y - actual_y)

```python
def calculate_error(predicted_Y, actual_Y):
    error = (predicted_Y - actual_Y)**2
    return error
```

```python
def cost_function(sum_error):
    cost = (sum_error)/(2*len(input_features))
    return cost
```

```python
def make_prediction(w, X, b):
    predicted_Y = w*X + b
    return predicted_Y
```

```python

#the model function will be Y = wX + b. so i will initialize w, b as 0
w, b, =  0, 0

learning_rate = .02
iteration = 500

for _ in range(0, iteration):
    global w, b

    #these variables will be used in gradient decent algorithm
    sum_of_partial_dervarive_w = 0
    sum_of_partial_dervarive_b = 0
    sum_error = 0
    
    for i in range(0, len(input_features)):
        predicted_Y = make_prediction(w, input_features[i], b)
        error = calculate_error(predicted_Y, output_features[i])
        sum_error += error

        #also calculating derivative that will be used in Gradient decent
        sum_of_partial_dervarive_w += (predicted_Y - output_features[i])*(input_features[i])
        sum_of_partial_dervarive_b += (predicted_Y - output_features[i])

    cost = cost_function(sum_error)
    
    #makeing Gradient decent
    temp_w = w - (learning_rate * sum_of_partial_dervarive_w)/len(input_features)
    temp_b = b - (learning_rate * sum_of_partial_dervarive_b)/len(input_features)
    w = temp_w
    b = temp_b

    if _ % 50 == 0:
        print(f"Iteration {_}: Cost = {cost:.4f}, w = {w:.4f}, b = {b:.4f}")
    
```
